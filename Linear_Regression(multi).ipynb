```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import r2_score,mean_squared_error
```

```
x,y=datasets.load_diabetes(True)
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=4)
```

```
lin_reg=LinearRegression()
lin_reg.fit(train_x,train_y)
lin_reg.coef_
```

```
pred=lin_reg.predict(test_x)
plt.plot(test_y,pred)
print("r2 :",r2_score(test_y,pred),"mse",mean_squared_error(test_y,pred))
```
