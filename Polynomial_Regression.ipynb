```
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
```

Make 2 degree polynomial regression.

```
x,y=datasets.load_diabetes(True)
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=4)
poly=PolynomialFeatures(2)
poly_train_x=poly.fit_transform(train_x)
poly_test_x=poly.fit_transform(test_x)
```

```
lin_reg=LinearRegression()
lin_reg.fit(poly_train_x,train_y)
```

This performs worse than the linear regression. Overfitting occurred.

```
pred=lin_reg.predict(poly_test_x)
plt.plot(test_y,pred)
print("r2 :",r2_score(test_y,pred),"mse",mean_squared_error(test_y,pred))
```
